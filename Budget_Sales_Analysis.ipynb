```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

### 1. Data Loading and Cleaning


```python
# Load the datasets
calendar_url = r"C:\Users\amani\Documents\++++Projects++++\ineuron_DA_Project\Dataset\AdventureWorks_Database.xlsx"
budget_url = r"C:\Users\amani\Documents\++++Projects++++\ineuron_DA_Project\Dataset\Budget.xlsx"
```

```python
# Read Excel sheets
Calender_Data = pd.read_excel(calendar_url, sheet_name="Calender")
Customer_Data = pd.read_excel(calendar_url, sheet_name="Customers")
Product_Data = pd.read_excel(calendar_url, sheet_name="Product")
Sales_Data = pd.read_excel(calendar_url, sheet_name="Sales")
Territory_Data = pd.read_excel(calendar_url, sheet_name="Territory")
Budget_Data = pd.read_excel(budget_url)
```

```python
# Display initial rows
print("Calendar Data:")
display(Calender_Data.head())

print("\nCustomer Data:")
display(Customer_Data.head())

print("\nProduct Data:")
display(Product_Data.head())

print("\nSales Data:")
display(Sales_Data.head())

print("\nTerritory Data:")
display(Territory_Data.head())

print("\nBudget Data (Initial):")
display(Budget_Data.head())
```

```python
# Clean Budget Data
Budget_Data = Budget_Data.iloc[2:, :]  # Remove first two metadata rows
Budget_Data.columns = Budget_Data.iloc[0]  # Set header from third row
Budget_Data = Budget_Data[1:].reset_index(drop=True)  # Remove header row and reset index
print("\nBudget Data (Cleaned):")
display(Budget_Data.head())
```

```python
# Drop irrelevant columns from Sales Data
columns_to_drop = ["Unnamed: 13","Unnamed: 14","Unnamed: 15","Unnamed: 16",
                  "Unnamed: 17","Unnamed: 18","Unnamed: 19","Unnamed: 22",
                  "StandardCost","List Price","diif std cost","diff list price"]
Sales_Data.drop(columns=columns_to_drop, axis=1, inplace=True)
print("\nSales Data (After Dropping Unnecessary Columns):")
display(Sales_Data.head())
```

```python
# Merge DataFrames
Merged_Data = pd.merge(Sales_Data, Customer_Data, on='CustomerKey', how='left')
Merged_Data = pd.merge(Merged_Data, Product_Data, on='ProductKey', how='left')
Merged_Data = pd.merge(Merged_Data, Territory_Data, on='SalesTerritoryKey', how='left')
print("\nMerged Data:")
display(Merged_Data.head())
```

```python
Merged_Data.shape[0]
```

```python
# Evaluate Missing Values
plt.figure(figsize=(18,6))
miss_merged_data = pd.DataFrame((Merged_Data.isnull().sum())*100/Merged_Data.shape[0]).reset_index()
miss_merged_data.columns = ['Column', 'Percentage']
sns.barplot(x='Column', y='Percentage', data=miss_merged_data)
plt.xticks(rotation=90, fontsize=9)
plt.title("Percentage of Missing Values in Merged Data")
plt.ylabel("Percentage")
plt.xlabel("Columns")
plt.show()
```

```python
miss_merged_data
```

```python
# Drop columns with more than 50% missing values
threshold = 40
cols_to_drop = miss_merged_data[miss_merged_data['Percentage'] > threshold]['Column'].tolist()
Merged_Data.drop(columns=cols_to_drop, inplace=True)
print(f"\nColumns dropped due to >{threshold}% missing values:", cols_to_drop)
```

```python
# Further Cleaning: Drop unuseful fields
columns_unuseful = ['ProductKey', 'PromotionKey','SalesTerritoryKey','SalesOrderNumber',
                    'SalesOrderLineNumber','TaxAmt','FirstName','LastName','FullName',
                    'AddressLine1','DaysToManufacture','ProductLine','Photo',
                    'ProductDescription','Region','Group','RegionImage']
Clean_Data = Merged_Data.drop(columns=columns_unuseful, axis=1)
print("\nCleaned Data Info:")
Clean_Data.info()
```

### 2. Initial Exploratory Analysis

```python
# 2.1 Gender Distribution of Customers
gender_counts = Clean_Data['Gender'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightpink'], startangle=90)
plt.title("Gender Distribution of Customers", fontsize=20)
plt.axis('equal')
plt.show()
```

```python
# 2.2 Marital Status Distribution
marital_status_counts = Clean_Data['MaritalStatus'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'], startangle=90)
plt.title("Marital Status Distribution of Customers", fontsize=20)
plt.axis('equal')
plt.show()
```

```python
# 2.3 Education Level of Customers
plt.figure(figsize=(10,6))
sns.countplot(x='Education', data=Clean_Data, order=Clean_Data['Education'].value_counts().index, palette='viridis')
plt.xticks(rotation=45)
plt.title('Customer Count by Education Level', fontsize=15)
plt.xlabel('Education Level')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# 2.5 Commute Distance vs Sales Amount
commute_sales = Clean_Data.groupby('CommuteDistance')['SalesAmount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=commute_sales.index, y=commute_sales.values, palette='coolwarm')
plt.xticks(rotation=45)
plt.title('Total Sales Amount by Commute Distance', fontsize=20)
plt.xlabel('Commute Distance')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# 2.6 Category Distribution of Customers
category_counts = Clean_Data['Category'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen'], startangle=90)
plt.title("Category Distribution of Customers", fontsize=20)
plt.axis('equal')
plt.show()
```

```python
# 2.7 SubCategory Distribution of Customers
plt.figure(figsize=(15,6))
sns.countplot(x='SubCategory', data=Clean_Data, order=Clean_Data['SubCategory'].value_counts().index, palette='Set2')
plt.xticks(rotation=45)
plt.title('Customer Count by SubCategory', fontsize=15)
plt.xlabel('SubCategory')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# 2.8 Country Distribution of Customers
country_counts = Clean_Data['Country'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title("Country Distribution of Customers", fontsize=20)
plt.axis('equal')
plt.show()
```

```python
# 2.9 Gender vs Sales Amount
gender_sales = Clean_Data.groupby('Gender')['SalesAmount'].sum()
plt.figure(figsize=(6,6))
gender_sales.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightpink'], startangle=90)
plt.title('Sales Distribution by Gender', fontsize=20)
plt.ylabel('')
plt.axis('equal')
plt.show()
```

```python
# 2.11 Correlation Heatmap
plt.figure(figsize=(18,9))
numeric_data = Clean_Data.select_dtypes(include='number')
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=20)
plt.show()
```

### 3. Data Insight Breakdown

```python
# 3.1 Sales Amount by Product Category
product_category_sales = Clean_Data.groupby('Category')['SalesAmount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=product_category_sales.index, y=product_category_sales.values, palette='Spectral')
plt.xticks(rotation=45)
plt.title('Total Sales Amount by Product Category', fontsize=20)
plt.xlabel('Product Category')
plt.ylabel('Sales Amount')
plt.show()

print('The highest sales amount is generated by Bikes category,')
```

```python
# 3.2 Monthly Sales Trend
Clean_Data['OrderDate'] = pd.to_datetime(Clean_Data['OrderDate'])
Clean_Data['MonthYear'] = Clean_Data['OrderDate'].dt.to_period('M')
monthly_sales_trend = Clean_Data.groupby('MonthYear')['SalesAmount'].sum()
plt.figure(figsize=(15,6))
monthly_sales_trend.plot(marker='o', color='teal')
plt.title('Monthly Sales Trend', fontsize=20)
plt.xlabel('Month-Year')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()

print("Sales peaked during dec-2016, indicating a seasonal or promotional impact on sales.")
```

```python
# 3.3 Product SubCategory Performance
sub_category_sales = Clean_Data.groupby('SubCategory')['SalesAmount'].sum().sort_values(ascending=False)
plt.figure(figsize=(16,6))
sns.barplot(x=sub_category_sales.index, y=sub_category_sales.values, palette='cool')
plt.xticks(rotation=45)
plt.title('Total Sales Amount by Product SubCategory', fontsize=20)
plt.xlabel('Product SubCategory')
plt.ylabel('Sales Amount')
plt.show()

print("Road Bikes is the leading subcategory, highlighting potential for targeted marketing.")
```

```python
# 3.4 Sales Performance by Year
Clean_Data['Year'] = Clean_Data['OrderDate'].dt.year
yearly_sales = Clean_Data.groupby('Year')['SalesAmount'].sum()
plt.figure(figsize=(10,6))
sns.lineplot(x=yearly_sales.index, y=yearly_sales.values, marker='o', color='orange')
plt.title('Yearly Sales Performance', fontsize=20)
plt.xlabel('Year')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()

print("Sales grew significantly in 2016, indicating strong business growth.")
```

```python
# 3.5 Analyzing Sales Amount vs. Quantity Sold
plt.figure(figsize=(12,6))
sns.scatterplot(x='OrderQuantity', y='SalesAmount', data=Clean_Data, hue='Category', palette='viridis')
plt.title('Sales Amount vs. Order Quantity by Category', fontsize=20)
plt.xlabel('Order Quantity')
plt.ylabel('Sales Amount')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

print('Higher order quantities generally lead to increased sales, especially within Bikes.')
```

```python
# 3.6 Average Sales Amount by Customer Education Level
education_sales_avg = Clean_Data.groupby('Education')['SalesAmount'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=education_sales_avg.index, y=education_sales_avg.values, palette='magma')
plt.xticks(rotation=45)
plt.title('Average Sales Amount by Education Level', fontsize=20)
plt.xlabel('Education Level')
plt.ylabel('Average Sales Amount')
plt.show()

print('Customers with Bachelors spend the most on average, suggesting targeted promotions.')
```

```python
# 3.7 Income Group Analysis
income_bins = [0, 30000, 60000, 90000, 120000, 150000]
income_labels = ['0-30k', '30k-60k', '60k-90k', '90k-120k', '120k+']
Clean_Data['Income_Group'] = pd.cut(Clean_Data['YearlyIncome'], bins=income_bins, labels=income_labels)
income_group_sales = Clean_Data.groupby('Income_Group')['SalesAmount'].sum()
plt.figure(figsize=(12,6))
sns.barplot(x=income_group_sales.index, y=income_group_sales.values, palette='inferno')
plt.title('Sales Amount by Income Group', fontsize=20)
plt.xlabel('Income Group')
plt.ylabel('Sales Amount')
plt.show()

# Insight: The 30k - 60k group contributes the most to sales, highlighting valuable customer segments.
```

```python
# 3.8 Average Sales Amount per Customer by Gender and Marital Status
avg_sales_gender_marital = Clean_Data.groupby(['Gender', 'MaritalStatus'])['SalesAmount'].mean().unstack()
plt.figure(figsize=(8,6))
avg_sales_gender_marital.plot(kind='bar', stacked=True, colormap='Pastel1')
plt.title('Average Sales Amount per Customer by Gender and Marital Status', fontsize=20)
plt.xlabel('Gender')
plt.ylabel('Average Sales Amount')
plt.legend(title='Marital Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
plt.show()

print('Female tend to spend more, which could influence personalized marketing strategies.')
```
